### Join Mortality and Group tables

```python
# Load Data from SQL database using SQL JOIN query
sql_join_mortality = '''SELECT group_table."GROUP_ID", "AREA", "RACE", "SEX", "SITE", "YEAR", "AGE_ADJUSTED_CI_LOWER", "AGE_ADJUSTED_CI_UPPER", "AGE_ADJUSTED_RATE", "EVENT_TYPE", "POPULATION", "CRUDE_CI_LOWER", "CRUDE_CI_UPPER", "CRUDE_RATE"
                        FROM group_table
                        RIGHT JOIN mortality_table ON group_table."GROUP_ID" = mortality_table."GROUP_ID"
                        '''
mortality_df2 = pd.read_sql(sql_join_mortality, con = engine)
mortality_df2.head()
```
